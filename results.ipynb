### Benchmarks results

```
import pandas as pd
```

```
df = pd.read_csv('results.log', delim_whitespace=True,
                 names=['bench', 'N', 'ns/op', 'ns/op_dup', 'B/op', 'B/op_dup', 'allocs/op', 'allocs/op_dup'],                 )
```

```
df = df.drop(['ns/op_dup', 'B/op_dup', 'allocs/op_dup'], axis=1)
```

```
df = df.drop([449, 450])
```

```
df['ns/op'] = df['ns/op'].astype(float)
df['B/op'] = df['B/op'].astype(float)
df['allocs/op'] = df['allocs/op'].astype(float)
```

```
def remove_bench_prefix(name: str, axis) -> str:
    return name.replace('Benchmark', '')

df['bench'] = df['bench'].apply(func=remove_bench_prefix, axis=1)
```

```
df['router'], df['bench'] = zip(*df['bench'].apply(lambda x: x.split('_', 1)))
```

```
df = df.sort_values('ns/op', axis=0)
```
